```python
df['NU_MEDIA_LP'].fillna(df.query("SG_UF_ESCOLA == 'SP'").groupby(["SG_UF_ESCOLA"])['NU_MEDIA_LP'].mean(), inplace=False)
plt.hist(df.query("SG_UF_ESCOLA == 'SP'"), bins=30)
```
